# merge

```python
df_merge1 = pd.read_csv("7-merge_data1.csv")
```

```python
df_merge1
```

```python
df_merge2 = pd.read_csv("7-merge_data2.csv")
```

```python
df_merge2
```

## merge - outer join

```python
df_merged_outer = pd.merge(df_merge1, df_merge2, on="Employee_ID", how="outer") 
```

```python
df_merged_outer #bütün dataları ele alır.
```

## merge - left join

```python
df_merged_left = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='left')
```

```python
df_merged_left #ilk tabloyu baz alır. ikinci tabloda, ona uygun olanları al gerisini göz ardı et
```

## merge - right join

```python
df_merged_right = pd.merge(df_merge1, df_merge2, on='Employee_ID', how='right')
```

```python
df_merged_right
```

## merge - inner join

```python
df_merged = pd.merge(df_merge1, df_merge2, on='Employee_ID', how="inner")
```

```python
df_merged # iki tablonun kesişim kümesini alır
```

```python
df = pd.read_csv('8-apply_function_data.csv')
```

```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["Salary_Category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
df
```

```python
def experience_category(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df["Adjusted"] = df["Experience"].apply(experience_category)
```

```python
df
```

```python
df["New_Score"] = df["Performance_Score"] + df["Adjusted"]
```

```python
df
```

```python
def adjust_new(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_Score"] = df.apply(adjust_new, axis = 1)
```

```python
df
```

```python
df["Formated_Name"] = df["Name"].apply(lambda x : x.replace("_", " "))
```

```python
df
```
