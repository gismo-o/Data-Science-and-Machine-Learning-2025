```python
numpy_arr1 = np.linspace(0,10,20) # 0'dan 10'a kadar 20 tane sayı dön ve arasını eşit tut
```

```python
numpy_arr1
```

```python
numpy_arr2 = numpy_arr1 ** 3 #üssünü al
```

```python
numpy_arr2
```

```python
plt.plot(numpy_arr1,numpy_arr2, "b*")
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2, "m*-")
plt.show()
```

```python
plt.plot(numpy_arr1,numpy_arr2, "r--")
plt.show()
```

```python
plt.subplot(1,2,1) #1 row, 2 kolon, 1.grafik.
# row -> 2si bir arada tek sırada, kolon -> 2 grafik
plt.plot(numpy_arr1, numpy_arr2, "r*-")

plt.subplot(1,2,2) #1 row, 2 kolon, 2.grafiği çiz
plt.plot(numpy_arr2, numpy_arr1, "g*-")
```

```python
my_figure = plt.figure()
figure_axes = my_figure.add_axes([0.1,0.1,0.8,0.4])  # sol-alt-genişlik-yükseklik yüzde olarak küçültüp büyültüyoruz
figure_axes.plot(numpy_arr1, numpy_arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph Title")
plt.show()
```

```python
new_fig = plt.figure (dpi = 200) #çözünürlük
new_axes = new_fig.add_axes([0.1,0.1,0.9,0.9])
new_axes.plot(numpy_arr1, numpy_arr1 ** 2, label="numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label="numpy array ** 3")
new_axes.legend(loc=10) #küçük bilgilendirme kutusu
plt.show()
```

# MATPLOTLIB STYLES

```python
data1 = np.linspace(0,10,20) # 0-10 arası 20 tane eşit dağılmış değer
```

```python
data1
```

```python
data2 = data1 ** 2
```

```python
# plt.subplots() #hazır figür ve eksen döndürür. tuple olarak döndürür
my_fig, my_axes =plt.subplots()
my_axes.plot(data1, data2, "#53CA0A", alpha=0.5)
my_axes.plot(data2, data1, "b")
plt.show()
```

```python
type(my_fig)
```

```python
type(my_axes)
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="blue", linewidth = 5.0) #çizgi kalınlığı
new_axes.plot(data1, data1 + 5, color="red", linewidth= 2.0)
new_axes.plot(data1, data1 + 6, color="brown", linestyle="-.")
new_axes.plot(data1, data1 + 7, color="black", linestyle=":")
new_axes.plot(data1, data1 + 8, color="green", linestyle="--")
new_axes.plot(data1, data1 + 9, color="yellow", linestyle="--", marker="o", markersize=8, markerfacecolor="red") #yuvarlarğın içini kırmızıyla doldur. boyutunu 8 yap 
new_axes.plot(data1, data1 + 10, color="gray", linestyle="--", marker="*", markersize=15)
new_axes.plot(data1, data1 + 12, color="gray", linestyle="--", marker="+", markersize=10)
plt.show()
```
