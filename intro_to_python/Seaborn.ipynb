# BOX PLOT

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#median, quartile (25%, 50%, 75%), min max 1.5 * IQR, outlier
```

```python
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100])

plt.figure(figsize = (6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(True)
plt.show()
```

```python
# dengeli veri
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60])

plt.figure(figsize = (6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(True)
plt.show()
```

```python
#dengesiz veri
data = np.array([5,7,9,15,20,22,25,30,32,35,37,40,50,55,60,100,2000])

plt.figure(figsize = (6,5))
sns.boxplot(y=data)
plt.title("Box Plot")
plt.ylabel("Data Value")
plt.grid(True)
plt.show()
```
